# Data Collection Overview

This notebook provides a basic overview of the TMDB and Trakt datasets using pandas.

```python
import pandas as pd
from pathlib import Path

# Setup paths
BASE_DIR = Path.cwd().parent.parent
TMDB_FILE = BASE_DIR / "data" / "movie_final_dataset.csv"
TRAKT_FILE = BASE_DIR / "data" / "trakt_ultimate_checkpoint.csv"
```

## 1. TMDB Dataset (movie_final_dataset.csv)

```python
tmdb_df = pd.read_csv(TMDB_FILE)
print("### TMDB Head ###")
display(tmdb_df.head())

print("\n### TMDB Info ###")
tmdb_df.info()

print("\n### TMDB Describe ###")
display(tmdb_df.describe())
```

## 2. Trakt Dataset (trakt_ultimate_checkpoint.csv)

```python
trakt_df = pd.read_csv(TRAKT_FILE)
print("### Trakt Head ###")
display(trakt_df.head())

print("\n### Trakt Info ###")
trakt_df.info()

print("\n### Trakt Describe ###")
display(trakt_df.describe())
```
